## feature engineering

```python
# Drop irrelevant or constant columns
df.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1, inplace=True)

# Binary encoding target
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# ✅ Feature Engineering Ideas

# 1. Total working years at company level
df['YearsAtCompanyLevel'] = df['YearsAtCompany'] / (df['TotalWorkingYears'] + 1)

# 2. Average tenure per company (avoid division by zero)
df['AvgTenurePerCompany'] = df['TotalWorkingYears'] / (df['NumCompaniesWorked'] + 1)

# 3. Years left to retirement (assuming 60 as retirement age)
df['YearsToRetire'] = 60 - df['Age']

# 4. MonthlyIncome per year (proxy for performance)
df['IncomePerYear'] = df['MonthlyIncome'] * 12 / (df['TotalWorkingYears'] + 1)

# 5. Satisfaction Score combining multiple indicators
df['SatisfactionLevel'] = df[['JobSatisfaction', 'EnvironmentSatisfaction', 'RelationshipSatisfaction']].mean(axis=1)

# 6. Overtime binary feature
df['OverTimeFlag'] = df['OverTime'].map({'Yes': 1, 'No': 0})

# 7. IsManager (based on JobRole)
df['IsManager'] = df['JobRole'].apply(lambda x: 1 if 'Manager' in x else 0)

# Show updated DataFrame sample
df.head()
```

```python
X= df.drop('Attrition', axis=1)
y = df['Attrition']

# Split columns
cat_cols = X.select_dtypes(include='object').columns.tolist()
num_cols = X.select_dtypes(include=np.number).columns.tolist()

# ColumnTransformer
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), num_cols),
    ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), cat_cols)
])

# SMOTE + Model Pipeline
model = ImbPipeline([
    ('pre', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('clf', LogisticRegression(max_iter=1000, class_weight='balanced'))
])

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Fit model
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
print("📊 Classification Report:\n", classification_report(y_test, y_pred))
print("🧾 Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt

# Get feature names after preprocessing
ohe = model.named_steps['pre'].named_transformers_['cat']
ohe_features = ohe.get_feature_names_out(cat_cols)
all_features = num_cols + list(ohe_features)

# Get coefficients
coef = model.named_steps['clf'].coef_[0]
feat_imp = pd.DataFrame({'Feature': all_features, 'Importance': coef}).sort_values(by='Importance', ascending=False)

# Plot
plt.figure(figsize=(10, 6))
feat_imp.set_index('Feature').head(15).plot(kind='barh', legend=False)
plt.title('Top 15 Feature Importances (Logistic Regression)')
plt.xlabel('Coefficient Value')
plt.tight_layout()
plt.show()
```
